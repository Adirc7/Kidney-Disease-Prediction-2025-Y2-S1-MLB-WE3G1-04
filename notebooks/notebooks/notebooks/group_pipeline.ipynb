```python
# === Full ML Pipeline: Members A–F ===
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, RobustScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Helper: consistent histogram + frequency polygon + mean/median
def plot_hist_freqpoly(series, title, xlabel):
    plt.figure(figsize=(6,4))
    counts, bins, _ = plt.hist(series.dropna(), bins=20, color="skyblue", edgecolor="black", alpha=0.7)
    mids = (bins[:-1] + bins[1:]) / 2
    plt.plot(mids, counts, color="green", marker="o", linestyle="-", linewidth=2, label="Frequency Polygon")
    mean_val = series.mean()
    median_val = series.median()
    plt.axvline(mean_val, color="red", linestyle="--", linewidth=2, label=f"Mean={mean_val:.2f}")
    plt.axvline(median_val, color="green", linestyle="-", linewidth=2, label=f"Median={median_val:.2f}")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel("Frequency")
    plt.legend()
    plt.tight_layout()
    plt.show()

# === Step 1: Load & Clean Column Names ===
df = pd.read_csv("C:\\Users\\TUF\\OneDrive\\Desktop\\SLIIT 2Y1S\\AI and ML\\Project\\AIML_Project\\kidney_disease_dataset(S).csv")
df.columns = df.columns.str.strip().str.lower().str.replace(r'[^\w]+', '_', regex=True)
print("\n[Step 1] Raw Data Loaded")
print("Shape:", df.shape)
print(df.head(), "\n")

# === Step 2: Select Useful Columns ===
useful_cols = [
    'age_of_the_patient','blood_pressure_mm_hg_','specific_gravity_of_urine','albumin_in_urine',
    'sugar_in_urine','random_blood_glucose_level_mg_dl_','blood_urea_mg_dl_','serum_creatinine_mg_dl_',
    'sodium_level_meq_l_','potassium_level_meq_l_','hemoglobin_level_gms_','packed_cell_volume_',
    'white_blood_cell_count_cells_cumm_','red_blood_cell_count_millions_cumm_',
    'estimated_glomerular_filtration_rate_egfr_','urine_protein_to_creatinine_ratio','urine_output_ml_day_',
    'serum_albumin_level','cholesterol_level','parathyroid_hormone_pth_level','serum_calcium_level',
    'serum_phosphate_level','body_mass_index_bmi_','cystatin_c_level','c_reactive_protein_crp_level',
    'interleukin_6_il_6_level','red_blood_cells_in_urine','pus_cells_in_urine','pus_cell_clumps_in_urine',
    'bacteria_in_urine','hypertension_yes_no_','diabetes_mellitus_yes_no_','coronary_artery_disease_yes_no_',
    'anemia_yes_no_','family_history_of_chronic_kidney_disease','urinary_sediment_microscopy_results','target'
]
df = df[useful_cols]
print("[Step 2] Selected Useful Columns")
print("Shape:", df.shape)
print(df.head(), "\n")

# === Member A: Missing Values & Cleaning ===
num_cols = df.select_dtypes(include=['float64','int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

missing_before = df.isna().sum().sum()
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
df = df.drop_duplicates()
missing_after = df.isna().sum().sum()
print("[Member A] Missing Values & Cleaning Done")
print(f"Missing values: before={missing_before}, after={missing_after}")
print("Shape:", df.shape, "\n")

# Visualization: Blood Pressure after cleaning
plot_hist_freqpoly(df["blood_pressure_mm_hg_"], "Blood Pressure After Cleaning", "Blood Pressure (mmHg)")

# === Member B: Categorical Encoding ===
le = LabelEncoder()
# Encode binary categoricals
for col in cat_cols:
    if col != 'target' and df[col].nunique() == 2:
        df[col] = le.fit_transform(df[col].astype(str))
# One-hot encode multi-category categoricals (exclude 'target')
multi_cat_cols = [col for col in cat_cols if col != 'target' and df[col].nunique() > 2]
df = pd.get_dummies(df, columns=multi_cat_cols, drop_first=True)
print("[Member B] Categorical Encoding Done")
print("Shape:", df.shape, "\n")

# === Member C: Outlier Handling (IQR) ===
# Choose the column to visualize
col_vis = "serum_creatinine_mg_dl_"

# Keep a copy BEFORE clipping
pre_out_creat = df[col_vis].copy()

# Clip all numeric columns using IQR
num_cols = df.select_dtypes(include=['float64','int64']).columns
for ncol in num_cols:
    Q1 = df[ncol].quantile(0.25)
    Q3 = df[ncol].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[ncol] = df[ncol].clip(lower, upper)

post_out_creat = df[col_vis].copy()
print("[Member C] Outlier Handling Done")
print("Shape:", df.shape, "\n")

# Boxplots with annotated quartiles/bounds (before & after)
def plot_box_with_bounds(series, title):
    Q1 = series.quantile(0.25)
    Q2 = series.quantile(0.50)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    plt.figure(figsize=(10,5))
    sns.boxplot(x=series, color="skyblue")
    for val, label, color in [
        (Q1, "Q1", "green"),
        (Q2, "Median (Q2)", "blue"),
        (Q3, "Q3", "green"),
        (lower, "Lower Bound", "red"),
        (upper, "Upper Bound", "red")
    ]:
        plt.axvline(val, color=color, linestyle="--", linewidth=1.5)
        plt.text(val, 0.95, f"{label}\n{val:.2f}", rotation=90,
                 va="top", ha="right", color=color, transform=plt.gca().get_xaxis_transform())
    plt.title(title)
    plt.xlabel(col_vis)
    plt.tight_layout()
    plt.show()

plot_box_with_bounds(pre_out_creat, "Serum Creatinine Before Outlier Handling")
plot_box_with_bounds(post_out_creat, "Serum Creatinine After Outlier Handling")

# === Member D: Scaling & Normalization ===
pre_scale_hgb = df["hemoglobin_level_gms_"].copy()
scaler = RobustScaler()
num_cols = df.select_dtypes(include=['float64','int64']).columns
df[num_cols] = scaler.fit_transform(df[num_cols])
print("[Member D] Scaling Done")
print("Shape:", df.shape, "\n")

# Visualization: Hemoglobin before vs after scaling
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
counts, bins, _ = plt.hist(pre_scale_hgb.dropna(), bins=20, color="skyblue", edgecolor="black", alpha=0.7)
plt.plot((bins[:-1]+bins[1:])/2, counts, color="green", marker="o", linestyle="-", linewidth=2, label="Frequency Polygon")
plt.axvline(pre_scale_hgb.mean(), color="red", linestyle="--", linewidth=2, label=f"Mean={pre_scale_hgb.mean():.2f}")
plt.axvline(pre_scale_hgb.median(), color="green", linestyle="-", linewidth=2, label=f"Median={pre_scale_hgb.median():.2f}")
plt.title("Hemoglobin Before Scaling")
plt.legend()

plt.subplot(1,2,2)
post_scale_hgb = df["hemoglobin_level_gms_"]
counts, bins, _ = plt.hist(post_scale_hgb.dropna(), bins=20, color="skyblue", edgecolor="black", alpha=0.7)
plt.plot((bins[:-1]+bins[1:])/2, counts, color="green", marker="o", linestyle="-", linewidth=2, label="Frequency Polygon")
plt.axvline(post_scale_hgb.mean(), color="red", linestyle="--", linewidth=2, label=f"Mean={post_scale_hgb.mean():.2f}")
plt.axvline(post_scale_hgb.median(), color="green", linestyle="-", linewidth=2, label=f"Median={post_scale_hgb.median():.2f}")
plt.title("Hemoglobin After Scaling (RobustScaler)")
plt.legend()
plt.tight_layout()
plt.show()

# === Member E: Feature Engineering ===
df["bmi_cat"] = pd.cut(df["body_mass_index_bmi_"], bins=[0,18.5,25,30,100],
                       labels=["Under","Normal","Over","Obese"])
df["age_group"] = pd.cut(df["age_of_the_patient"], bins=[0,30,50,70,200],
                         labels=["Young","Adult","Mid","Elder"])
for col in ["bmi_cat","age_group"]:
    df[col] = le.fit_transform(df[col].astype(str))
print("[Member E] Feature Engineering Done")
print("Shape:", df.shape, "\n")

# === Member F: Multi-Class Target Handling (Proportional Oversampling) ===
X = df.drop(columns=['target'])
y = df['target']

print("[Member F] Target Prepared")
print("Shape X:", X.shape, "Shape y:", y.shape)
print("Target Distribution:\n", y.value_counts(), "\n")

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

print("Class distribution before balancing:\n", y_train.value_counts(), "\n")

# Build proportional sampling strategy: sum of disease subtypes equals No_Disease
majority_count = y_train.value_counts()['No_Disease']
disease_subs = ['Low_Risk', 'Moderate_Risk', 'High_Risk', 'Severe_Disease']  # include High_Risk
sub_counts = y_train.value_counts()[disease_subs]
sub_total = sub_counts.sum()
sub_proportions = sub_counts / sub_total
target_sub_counts = (sub_proportions * majority_count).round().astype(int)

sampling_strategy = {'No_Disease': majority_count}
sampling_strategy.update(target_sub_counts.to_dict())

print("Target sampling strategy:\n", sampling_strategy, "\n")

# Apply RandomOverSampler
ros = RandomOverSampler(sampling_strategy=sampling_strategy, random_state=42)
X_train_bal, y_train_bal = ros.fit_resample(X_train, y_train)

print("Class distribution after balancing:\n", pd.Series(y_train_bal).value_counts(), "\n")

# Visualization: Before vs After Balancing (fixed class order)
class_order = ['No_Disease','Low_Risk','Moderate_Risk','High_Risk','Severe_Disease']
fig, ax = plt.subplots(1,2, figsize=(12,5))
y_train.value_counts().reindex(class_order).plot(kind='bar', ax=ax[0], color="red", title="Before Balancing")
pd.Series(y_train_bal).value_counts().reindex(class_order).plot(kind='bar', ax=ax[1], color="green", title="After Balancing (Proportional)")
for a in ax:
    a.set_xlabel("Class")
    a.set_ylabel("Count")
plt.tight_layout()
plt.show()

# === Simple Model Training (Multi-Class Logistic Regression) ===
model = LogisticRegression(max_iter=2000, multi_class="multinomial", solver="lbfgs")
model.fit(X_train_bal, y_train_bal)
y_pred = model.predict(X_test)

print("[Final Step] Logistic Regression Results")
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Final dataset info
print("\nFinal DataFrame Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("\nSample rows:")
df.head(13145)
```
